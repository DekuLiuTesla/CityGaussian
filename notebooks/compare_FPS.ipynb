```python
import os
import sys
import torch
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
heights = [5, 10, 15, 20, 25, 30]
Mean_Org = [22.0, 18.1, 15.1, 13.7, 13.4, 13.2]
Mean_FPS_LoD_2 = [57.82, 43.54, 38.00, 34.18, 30.34, 20.56]
Mean_FPS_LoD_1 = [62.30, 51.30, 43.17, 37.23, 33.22, 30.42]
Mean_FPS_LoD = [66.56, 55.82, 47.87, 42.59, 39.09, 36.37]

Min_Org = [16.5, 13.7, 11.6, 9.73, 9.41, 9.49]
Min_FPS_LoD_2 = [37.2, 30.3, 22.9, 19.6, 17.7, 8.4]
Min_FPS_LoD_1 = [45.54, 35.25, 32.23, 26.69, 24.24, 20.66]
Min_FPS_LoD = [39.4, 38.5, 34.0, 29.6, 28.8, 27.8]
```

```python
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(16, 12))
size1, size2, line_width = 15, 20, 3
plt.plot(heights, Mean_Org, label='Original', color='gray', marker='s', markersize=size1, linewidth=line_width)
plt.plot(heights, Mean_FPS_LoD_2, label='LoD 2 Only', color=plt.get_cmap('Paired')(3), marker='D', markersize=size1, linewidth=line_width)
plt.plot(heights, Mean_FPS_LoD_1, label='LoD 1 Only', color=plt.get_cmap('Paired')(1), marker='o', markersize=size1, linewidth=line_width)
plt.plot(heights, Mean_FPS_LoD, label='LoD', color=plt.get_cmap('Accent')(5), marker='*', markersize=size2, linewidth=line_width)

plt.fill_between(heights, Min_Org, Mean_Org, color='black', alpha=0.1)
plt.fill_between(heights, Min_FPS_LoD_2, Mean_FPS_LoD_2, color='g', alpha=0.1)
plt.fill_between(heights, Min_FPS_LoD_1, Mean_FPS_LoD_1, color='b', alpha=0.1)
plt.fill_between(heights, Min_FPS_LoD, Mean_FPS_LoD, color='r', alpha=0.1)

# plt.plot(heights, Min_Org, color='gray', marker='s', linestyle='dashed', markersize=size1, linewidth=line_width)
# plt.plot(heights, Min_FPS_LoD_2, color='limegreen', marker='D', linestyle='dashed', markersize=size1, linewidth=line_width)
# plt.plot(heights, Min_FPS_LoD_1, color='dodgerblue', marker='o', linestyle='dashed', markersize=size1, linewidth=line_width)
# plt.plot(heights, Min_FPS_LoD, color='red', marker='*', linestyle='dashed', markersize=size2, linewidth=line_width)

font_size = 36
plt.xticks(fontsize=font_size)
plt.yticks(fontsize=font_size)
plt.xlabel('Camera Height / 100m', fontsize=font_size)
plt.ylabel('FPS', fontsize=font_size)
plt.legend(fontsize=font_size-8)
plt.grid(False)
```
